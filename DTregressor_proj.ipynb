```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from Decision_Tree_Code import DecisionTreeREG
from sklearn.tree import DecisionTreeRegressor
```

```python
data = pd.read_csv("data.csv")
data
```

```python
x = data.drop("y", axis=1).values
y = data["y"].values
```

```python
model = DecisionTreeREG(max_depth=10)
model.fit(x,y)
```

```python
plt.scatter(x, y)
y_prediction = model.predict(x)
plt.plot(x, y_prediction, c="r")
```

```python
model = DecisionTreeRegressor()
model.fit(x, y)
```

```python
plt.scatter(x, y)
y_prediction = model.predict(x)
plt.plot(x, y_prediction, c="r")
```
